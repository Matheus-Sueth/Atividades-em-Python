```python
def imprimir(numero):
  for i in range(1,numero+1):
    for _ in range(i):
      print(f'{i: >3}', end='  ')
    print()

imprimir(10)
```

#image.png

```python
def imprimir2(numero):
  for i in range(1,numero+1):
    for j in range(1,i+1):
      print(f'{j: >3}', end='  ')
    print()

imprimir2(3)
```

```python
def soma(n1, n2, n3):
  return n1 + n2 + n3

print(soma(4,5,6))
print(soma(9,8,7))
```

```python
def valorpn(valor):
  if valor > 0:
    return 'P'
  else:
    return 'N'

print(valorpn(1))
print(valorpn(0))
print(valorpn(-1))
```
