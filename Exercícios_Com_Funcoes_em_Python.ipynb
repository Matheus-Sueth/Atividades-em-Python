```python
def imprimir(numero):
  for i in range(1,numero+1):
    for _ in range(i):
      print(f'{i: >3}', end='  ')
    print()

imprimir(10)
```

```python
def conversaoHora(hora,minutos):
  if hora < 1 or hora > 24 or minutos < 1 or minutos > 60:
    return 'Erro'
  else:
    return saidaHora(hora,minutos)

def saidaHora(hora,minutos):
  if hora > 12:
    hora = hora % 12
    return f'{hora:02}:{minutos:02} P.M.'
  else:
    return f'{hora:02}:{minutos:02} A.M.'

print('Digite 0 para sair')
while True:
  try:
    hora, minuto = input().split(':')
    print(conversaoHora(int(hora),int(minuto)),'\n')
  except:
    break
```

```python
def valorPagamento(vlPrestacao, nrDias):
  if int(nrDias) == 0:
    return vlPrestacao
  else:
    vlPrestacao += (vlPrestacao * 0.03)
    vlPrestacao += (vlPrestacao * (nrDias/100))
    return vlPrestacao

totalPrestacoes = []

print('Digite 0 para sair\n')
while True:
  valor = input('Digite o valor da presta√ß√£o: ')
  if valor == '0':
    break
  else:
    numero = input('Digite o n√∫mero de dias em atraso: ')
    valorTotal = valorPagamento(float(valor),float(numero))
    totalPrestacoes.append(valorTotal)
    print('R$',valorTotal,'\n')

print(f'\n\nA quantidade de presta√ß√µes = {len(totalPrestacoes)}\nO valor total de presta√ß√µes pagas no dia = R$ {sum(totalPrestacoes)}')
```

```python
def reverso(numero):
  string = str(numero)
  return string[::-1]

def reverso2(numero):
  string = str(numero)
  aux = ''
  for i in range(len(string)-1,-1,-1):
    aux+=string[i]
  return aux

print(reverso(1000))
print(reverso2(973823))
```

```python
def craps():
  dados = randint(2,12)
  return dados

def loop(ponto):
  while True:
    resultado = craps()
    if resultado == 7:
      print('Voc√™ perdeu')
      break
    if resultado == ponto:
      print('Voc√™ ganhou')
      break

def inicio():
  resultado = craps()
  if resultado == 7 or resultado == 11:
    print('Voc√™ √© natural e ganhou')
  elif resultado == 2 or resultado == 3 or resultado == 12:
    print('CRAPS e voc√™ perdeu')
  else:
    print(f'Ponto = {resultado}')
    loop(resultado)

inicio()
```

```python
def dataExtenso(data):
  try:
    dt = data.split('/')
    if int(dt[0]) < 1 or int(dt[0]) > 31 or int(dt[2]) < 0:
      return 'NULL'
    return f'{int(dt[0])} de {meses[int(dt[1])]} de {int(dt[2])}'
  except:
    return 'NULL'

meses = {1:'Janeiro', 2:'Fevereiro', 3:'Mar√ßo', 4:'Abril', 5:'Maio', 6:'Junho', 7:'Julho',
         8:'Agosto', 9:'Setembro', 10:'Outubro', 11:'Novembro', 12:'Dezembro'}

print(dataExtenso('7/10/2021'))
```

```python
def embaralha(palavra):
  lista = []
  nvPalavra = ''
  tamanho = len(palavra)
  while True:
    i = randint(0,tamanho-1)
    if i in lista:
      continue
    else:
      lista.append(i)
      nvPalavra+=palavra[i]
    if len(lista) == tamanho:
      return nvPalavra

print(embaralha('python'))
```
