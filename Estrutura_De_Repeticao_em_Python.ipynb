```python
import numpy as np
from random import uniform

codigo = []
altura = []
peso = []
i = 10
while not codigo.count(0):
  codigo.append(i)
  if codigo.count(0):
    codigo.remove(0)
    break
  altura.append(round(uniform(1, 2),2))
  peso.append(round(uniform(60, 150),2))
  i -= 1

i = altura.index(max(altura))
print('Cliente mais alto')
print(f'Código: {codigo[i]}')
print(f'Altura: {altura[i]}m')
print(f'Peso: {peso[i]}Kg')
i = altura.index(min(altura))
print('Cliente mais baixo')
print(f'Código: {codigo[i]}')
print(f'Altura: {altura[i]}m')
print(f'Peso: {peso[i]}Kg')
i = peso.index(max(peso))
print('Cliente mais pesado')
print(f'Código: {codigo[i]}')
print(f'Altura: {altura[i]}m')
print(f'Peso: {peso[i]}Kg')
i = peso.index(min(peso))
print('Cliente mais leve')
print(f'Código: {codigo[i]}')
print(f'Altura: {altura[i]}m')
print(f'Peso: {peso[i]}Kg')

print(f'Média das alturas: {round(np.mean(altura),2)}m')
print(f'Média dos pesos: {round(np.mean(peso),2)}Kg')
```

```python
from random import randint
import numpy as np

gabarito = []
questoes = {0:'A', 1:'B', 2:'C', 3:'D', 4:'E'}
for nota_gabarito in range(1,11):
  gabarito.append(questoes[randint(0,4)])
print('Professor adicionou o gabarito')
alunos = {}
nome_alunos = {}
while True:
  nome = input('Qual o seu nome: ')
  r_questao = []
  for _ in range(1,len(gabarito)+1):
    r_questao.append(questoes[randint(0,4)])
  alunos[nome] = r_questao
  resposta = input('O próximo vai utilizar o sistema ?').lower()
  if resposta == 'n' or resposta == 'nao' or resposta == 'não':
    break

nota = {}
for nomes in alunos.keys():
  acertos = 0
  erros = 0
  for indice in range(0,len(gabarito)):
    if alunos[nomes][indice] == gabarito[indice]:
      acertos+=1
    else:
      erros+=1
  nota[nomes] = [acertos, erros]
maior_nota = 0
menor_nota = 0
todas_notas = 0
for aluno in nota.keys():
  if nota[aluno][0] > maior_nota:
    maior_nota = nota[aluno][0]
  if nota[aluno][1] < menor_nota:
    menor_nota = nota[aluno][1]
  todas_notas+=nota[aluno][0]

print('Maior acerto = ', maior_nota)
print('Menor acerto = ', menor_nota)
print('O total de alunos foi: ', len(nota.keys()))
print('Média do alunos = ', np.mean(todas_notas))
```

```python
numero = 5
m = 1
soma = 0
print('S = ',end='')
for n in range(1,numero+1):
  print(f'{n}/{m}',end='')
  soma+=(n/m)
  if n == numero:
    break
  print(' + ', end='')
  m+=2
print(f'\nSoma = {soma: .3f}')
```
