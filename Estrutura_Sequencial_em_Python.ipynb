```python
numeroHorasTrabalhadasMes = []

ganhaHora = float(input('Digite o ganho por hora: '))
numeroHorasTrabalhadasMes = input('Digite o número de horas trabalhadas: Exemplo 01:30 ').split(":")

horasporMes = int(numeroHorasTrabalhadasMes[1]) / 60
salario_Bruto = (ganhaHora * int(numeroHorasTrabalhadasMes[0])) + (ganhaHora*horasporMes)
print('+ Salário Bruto : R$ ' + str(round(salario_Bruto,2)))
iR = salario_Bruto * 0.11
salario_Bruto = salario_Bruto - iR
print('- IR (11%) : R$ ' + str(round(iR,2)))
inss = salario_Bruto * 0.08
salario_Bruto = salario_Bruto - inss
print('- INSS (8%) : R$ ' + str(round(inss,2)))
sindicato = salario_Bruto * 0.05
salario_Bruto = salario_Bruto - sindicato
print('- Sindicato (5%) : R$ ' + str(round(sindicato,2)))
print('= Salário Líquido : R$ ' + str(round(salario_Bruto,2)))
```

#image.png

```python
tamanho = float(input('Quantos metros quadrados você precisa: '))
preço_lata = 80
latalitros = 18
preço_galao = 25
galaolitros = 3.6

q_Litros = tamanho / 6
latas = galao = 0


print('Digite: a = lata')
print('Digite: b = galao')
print('Digite: c = galao e lata')
situacao = input()

if (situacao == 'a'):
  while (q_Litros > 0):
    latas += 1
    q_Litros -= latalitros
  q_final = round(q_Litros,2)*-1
  preco_Total = latas * preço_lata
  if (q_final > 0):
    print('Lata = ' + str(latas))
    print('Sobrou = ' + str(q_final))
    print('Preço = R$ ' + str(preco_Total))
  else:
    print('Lata = ' + str(latas))
    print('Sobrou Nada')
    print('Preço = R$ ' + str(preco_Total))

elif (situacao == 'b'):
  while (q_Litros > 0):
    galao += 1
    q_Litros -= galaolitros
  q_final = round(q_Litros,2)*-1
  preco_Total = galao * preço_galao
  if (q_final > 0):
    print('Galão = ' + str(galao))
    print('Sobrou = ' + str(q_final))
    print('Preço = R$ ' + str(preco_Total))
  else:
    print('Galão = ' + str(galao))
    print('Sobrou Nada')
    print('Preço = R$ ' + str(preco_Total))

elif (situacao == 'c'):
    while (q_Litros > 0):
      if (q_Litros >= 18):
        latas += 1
        q_Litros -= latalitros
      else:
        galao += 1
        q_Litros -= galaolitros
    q_final = round(q_Litros,2)*-1
    preco_Total = (galao * preço_galao) + (latas * preço_lata)
    if (q_final > 0):
      print('Lata = ' + str(latas))
      print('Galão = ' + str(galao))
      print('Sobrou = ' + str(q_final))
      print('Preço = R$ ' + str(preco_Total))
    else:
      print('Lata = ' + str(latas))
      print('Galão = ' + str(galao))
      print('Sobrou Nada')
      print('Preço = R$ ' + str(preco_Total))

else:
  print('ERRO !!!')
```
