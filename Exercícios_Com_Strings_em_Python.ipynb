```python
def conteudo(str1, str2):
  if len(str1) == len(str2):
    for i in range(len(str1)):
      if str1[i] != str2[i]:
        return 'As duas strings possuem conte√∫dos diferentes.'
    return 'As duas strings possuem conte√∫dos iguais.'

print('Compara duas strings')
str1 = input('String 1: ')
str2 = input('String 2: ')
tam1 = len(str1)
tam2 = len(str2)
print(f'Tamanho de "{str1}": {tam1} caracteres')
print(f'Tamanho de "{str2}": {tam2} caracteres')
if tam1 == tam2:
  print('As duas strings tem os tamanhos iguais.')
  print(conteudo(str1,str2))
else:
  print('As duas strings s√£o de tamanhos diferentes.')
  print('As duas strings possuem conte√∫dos diferentes')
```
