```python
def palindromo(palavra):
  aux2 = ''
  for i in palavra:
    if i == ' ':
      continue
    else:
      aux2+=i
  aux = aux2[::-1]
  for i in range(len(aux)):
    if aux[i] != aux2[i]:
      return 'N√£o √© pal√≠ndromo'  
  return '√â pal√≠ndromo'


print(palindromo('OSSO'))
print(palindromo('DEDO'))
print(palindromo('SUBI NO ONIBUS'))
print(palindromo('ATIVAR O WINDOWS'))
```

```python
def validarDigito(j, cpf):
  soma = 0
  for i in range(len(cpf)):
    if j > 1:
      soma+=int(cpf[i])*j
      j-=1
  digito = 11 - (soma % 11)
  if digito >= 10:
    digito-=10
  return digito

def verificaNumero(cpf):
  for i in range(10):
    if cpf.count(i) == 11:
      return False
  return True  

def verificaCPF(cpf):
  if cpf.count('.') != 2 and cpf.count('-') != 1:
    return 'Erro na pontua√ß√£o'
  if len(cpf) != 14:
    return 'Erro na quantidade de d√≠gitos'
  cpf = re.split('-|\.|', cpf)
  while True:
    try:
      cpf.remove('')
    except:
      break
  if verificaNumero(cpf):
    return 'CPF inv√°lido'
  digito = validarDigito(10, cpf)
  if str(digito) != cpf[9]:
    return 'Primeiro d√≠gito inv√°lido'
  digito = validarDigito(11, cpf)
  if str(digito) != cpf[10]:
    return 'Segundo d√≠gito inv√°lido'
  return 'CPF v√°lido'

cpf = ['111.111.111-11','123.456.789-00','941.623.262-276','723.820.237.76','199.704.338-65']
for n in cpf:
  print(verificaCPF(n))
```

##Fun√ß√£o para embaralhar

```python
def embaralha(palavra):
  lista = []
  nvPalavra = ''
  tamanho = len(palavra)
  while True:
    i = randint(0,tamanho-1)
    if i in lista:
      continue
    else:
      lista.append(i)
      nvPalavra+=palavra[i]
    if len(lista) == tamanho:
      return nvPalavra
```
