##ABRE O ARQUIVO COM OS IPs E GERA OUTRO ARQUIVO COM OS IPs validados

```python
def validar(ip):
  cont = 0
  for n in ip:
    if int(n) > 255 or int(n) < 0:
      cont+=1
  if cont == 0:
    return True
  return False

arquivo = open('endereco.txt','r')
auxiliar = open('validacao.txt','w')
listaIp = arquivo.readlines()
listaIpValido = []
listaIpInvalido = []
for ip in listaIp:
  listaIp[listaIp.index(ip)] = ip[:-1]
for ip in listaIp:
  ipp = ip.split('.')
  teste = validar(ipp)
  if teste:
    listaIpValido.append(ip)
  else:
    listaIpInvalido.append(ip)

auxiliar.write('[Endere√ßos v√°lidos:]\n')
for ip in listaIpValido:
  auxiliar.write(ip+'\n')

auxiliar.write('\n[Endere√ßos inv√°lidos:]\n')
for ip in listaIpInvalido:
  auxiliar.write(ip+'\n')

arquivo.close()
auxiliar.close()
```
